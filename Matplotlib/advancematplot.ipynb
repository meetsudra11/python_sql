```python
iris = pd.read_csv('iris.csv')
iris.head(5)
```

```python
# scatter plot shows correlation amoung variables 
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(3)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)#winter,jet,summer,viridis, # alpha value gives the obesity to the chart 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar() # values dega 
```

```python

plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)#winter,jet,summer,viridis, # alpha value gives the obesity to the chart 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar() # values dega 
```

```python
bat = pd.read_csv('batter.csv')
bat.head()
```

```python
df_sample = bat.head(100).sample(25,random_state=5)
df_sample
```

```python
plt.scatter(df_sample['avg'],df_sample['strike_rate'],s=df_sample['runs'])

for i in range(df_sample.shape[0]):
    plt.text(df_sample['avg'].values[i],df_sample['strike_rate'].values[i],df_sample['batter'].values[i])
```

```python
# players with strike rate greater then 130 and avg greater then 30
plt.scatter(df_sample['avg'],df_sample['strike_rate'])

plt.axhline(130,color='red')#axis horizontal line
plt.axhline(140,color='green')# exceptional hitters with high strike rate 
plt.axvline(30,color='red')


for i in range(df_sample.shape[0]):
    plt.text(df_sample['avg'].values[i],df_sample['strike_rate'].values[i],df_sample['batter'].values[i])
```

```python
# A different way to plot graphs 
bat.head(3)
```

```python
plt.figure(figsize=(15,6))
plt.scatter(bat['avg'],bat['strike_rate'],color='green')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(bat['avg'],bat['strike_rate'],color='green')

ax.set_title('avg and strike rate of players')
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')

fig.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(bat['avg'],bat['runs'])
ax[0].set_title('avg n runs')
ax[0].set_ylabel('runs')

ax[1].scatter(bat['avg'],bat['strike_rate'])
ax[1].set_title('avg n strike_rate')
ax[1].set_xlabel('average')
ax[1].set_ylabel('strike_rate')


plt.show()
```

```python
# another way 
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)# 2rows 1column 1stchart 
ax1.scatter(bat['avg'],bat['strike_rate'])

ax2 = fig.add_subplot(2,1,2)# 2rows 1 column 2ndchart
ax2.scatter(bat['avg'],bat['runs'])
```
