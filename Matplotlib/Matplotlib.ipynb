```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed',linewidth='6') # solid, dashed, dotted, dashdot
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed', marker='+')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
# legend 
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed',label='Rohit')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed',label='Kohli')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')

plt.legend() # loc='best' it fits in the best available space 
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed', marker='+')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')

plt.legend()
plt.grid()

plt.show()  # this enables the graph in few text editors 
```

```python
batter = pd.read_csv('batter.csv')
batter = batter.head(50)
x = batter['avg']
y = batter['strike_rate']
```

```python
plt.scatter(x,y,color='red',marker='+')
plt.title('Average and SR analysis of Top 50 batsman')
plt.xlabel('AVERAGE')
plt.ylabel('STRIKE-RATE')

plt.grid()
plt.show()
```

higer average means how reliable is the player whereas strike_rate shows how destructive is the player

# Barchart

Bivariate analysis and uni as well 

Numerical vs Categorical

Aggregate analysis of groups 


```python
children = [20,25,33,20,25]
color = ['Red','Blue','Green','Yellow','Pink']

plt.bar(color,children,color='green')
```

```python

plt.barh(color,children,color='green')
```

```python
batsman = pd.read_csv('batsman_season_record.csv')
batsman
```

```python
plt.bar(batsman['batsman'],batsman['2015'],width=0.5)
```

```python
batsman
```

```python
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=0.20)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=0.25)

plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2017'],width=0.20)

plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])

plt.show()
```

```python
# xticks problem 

# names in  the x axis overlap at times there we use 
plt.bar(np.arange(batsman.shape[0])-0.2,batsman['2015'],width=0.20)
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=0.25)

plt.bar(np.arange(batsman.shape[0])+0.2,batsman['2017'],width=0.20)

plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'],rotation=('vertical'))

plt.show()
```

```python
batsman
```

## Stacked Bar chart

```python
plt.bar(batsman['batsman'],batsman['2017'],label='2017')

plt.bar(batsman['batsman'],batsman['2016'],bottom=batsman['2017'],label='2016')

plt.bar(batsman['batsman'],batsman['2015'],bottom=(batsman['2016']+batsman['2017']),label='2015')

plt.legend()
```

# Piechart

Contribution on total

```python
data = [23,45,100,20,49]
sub=['math','eng','hindi','sanskrit','hist']
plt.pie(data,labels=sub)
plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','green','blue','violet','purple','yellow'])

plt.show()
```

```python
# removing a particular section

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0,0,0,0,0.1,0],shadow=True) # shadow enhances the graph

plt.show()
```
