```python
# dataframe 
rs_vk = pd.read_csv('sharma-kohli.csv')
rs_vk
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'])
plt.plot(rs_vk['index'],rs_vk['V Kohli'])
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed',linewidth='6') # solid, dashed, dotted, dashdot
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed', marker='+')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
# legend 
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed',label='Rohit')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed',label='Kohli')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')

plt.legend() # loc='best' it fits in the best available space 
```

```python
plt.plot(rs_vk['index'],rs_vk['RG Sharma'],color='red',linestyle='dashed', marker='+')
plt.plot(rs_vk['index'],rs_vk['V Kohli'],color='green',linestyle='dashed')
plt.title('Ro-hit and chiku bhai comparison')
plt.xlabel('Season')
plt.ylabel('Score')

plt.legend()
plt.grid()

plt.show()  # this enables the graph in few text editors 
```

```python
# limiting axes 
price = [48000,55000,75000,60000,61000,5000000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.ylim(1,100000)
plt.xlim(2015,2019)
```

# Scatter Plot

for finding correlation between two entities 

```python
x = np.linspace(-10,10,50)
x

y =10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
batter = pd.read_csv('batter.csv')
batter = batter.head(50)
x = batter['avg']
y = batter['strike_rate']
```

```python
plt.scatter(x,y,color='red',marker='+')
plt.title('Average and SR analysis of Top 50 batsman')
plt.xlabel('AVERAGE')
plt.ylabel('STRIKE-RATE')

plt.grid()
plt.show()
```

higer average means how reliable is the player whereas strike_rate shows how destructive is the player
